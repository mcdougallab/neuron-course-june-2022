# Load the morphology into NEURON

````python
from neuron import h
h.load_file('import3d.hoc')

# load the morphology and instantiate at the top level (i.e. not in a class)
cell = h.Import3d_SWC_read()
cell.input('c91662.CNG.swc')
h.Import3d_GUI(cell, 0)
i3d = h.Import3d_GUI(cell, 0)
i3d.instantiate(None)  # pass in an object to instantiate inside the object instead
````

# Explore it

At this point, there are a number of sections. If we had passed in a object, the sections would be created inside the object and there would be Python lists of sections within regions (`ob.dend`, ...). Since we didn't do that, the sections were placed inside `h`:

````python
h.apic[37].L
````

````python
sec = h.apic[37]
for i in range(sec.n3d()):
  print(f"({sec.x3d(i)}, {sec.y3d(i)}, {sec.z3d(i)}; {sec.diam3d(i)})")
````

````python
print(sec.parentseg())
````

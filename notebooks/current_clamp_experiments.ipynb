```python
from neuron import h
from neuron.units import um, mV, ms
h.load_file("stdrun.hoc")

import plotly
```

```python
axon = h.Section(name="axon")
```

```python
axon.L = 10001 * um
axon.diam = 1 * um
```

```python
axon.nseg = 1001
```

```python
h.hh.insert(axon)
```

```python
iclamp = h.IClamp(axon(0))
iclamp.delay = 1 * ms
iclamp.dur = 0.1
iclamp.amp = 1
```

```python
rvp = h.RangeVarPlot("v", axon(0), axon(1))
```

```python
axon.nseg = 1001
```

```python
h.finitialize(-65 * mV)
```

```python
my_plot = rvp.plot(plotly, name="t=0", line={"width": 4})
```

```python
my_plot.show()
```

```python
for tstop in [0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5, 10]:
  h.continuerun(tstop)
  my_plot = rvp.plot(my_plot, name=f"t={tstop}", line={"width": 4})
```

```python
my_plot = my_plot.update_layout({
    "xaxis_title": "position (µm)",
    "yaxis_title": "membrane potential (mV)"
})
```

```python
my_plot.show()
```
